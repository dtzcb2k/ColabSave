```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches # 提供不同的形狀
```

```python
plt.get_backend()
```

```python
# 繪圖純粹用list方式
# 參考教學網站 https://simplelearn.tw/matplotlib-intro/
x = [5, 6, 7, 8, 9, 8, 7, 6, 5]
y = [2, 1.5, 1.2, 1.5, 2, 2.3, 2, 2.3, 2]
z=[2.5,0,2,1,3]
# 簡易的變換線條顏色和線粗、節點圖案變化("o","x",".")、線條類型、線條名稱
plt.plot(x, y, color='red', linewidth=0.5, marker="x", linestyle="--", label="heart")
plt.plot(z, label="z")
plt.plot(y, color='yellow', linewidth=5, marker="o", linestyle="-", label="y")
# 對點進行標註("訊息",在哪一點，文字顯示在何處，顏色)為了知道是指哪一點可以加上箭頭
plt.annotate("heart", xy=(7,2), xytext=(6.64,1.5), color="red", arrowprops={"color":"black"})

# label要是沒有用legend並指定位置了話不會顯示
plt.legend(loc="upper left")# loc是local也就是位置
plt.xlabel('x.aix', color="green")
plt.ylabel('y.aix')
plt.title("純粹用list方式")
plt.show()

# 補: xticks([1,2,3],[a,b,c])置換x軸5中的123變成abc
```

# 繪製貓咪

使用到GPT協助
輸入的指令為 該如何用matplotlib繪製貓咪圖案，詳細解釋每段程式碼
https://chatgpt.com/share/67becadb-3780-800f-b1d1-78cab1838817

```python
# 創建畫布，fig是整個畫布，而ax則是座標軸，有ax我們才能在該區域進行繪製
fig, ax = plt.subplots(figsize=(6, 6))
# 限制xy軸範圍，或說設置畫面大小
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
ax.axis('off')  # 隱藏座標軸

# 1. 繪製貓咪的頭部（圓形）((圓心),半徑,顏色(不是用color去設置))
head = patches.Circle((0, 0), 1, facecolor='lightgray', edgecolor='black', linewidth=2)
ax.add_patch(head)

# 2. 繪製耳朵（兩個三角形）(定義三角形的三點位置)
left_ear = patches.Polygon([[-0.8, 0.6], [-0.8, 1.2], [0, 1]], closed=True, facecolor='gray', edgecolor='black', linewidth=2)
right_ear = patches.Polygon([[0.8, 0.6], [0.8, 1.2], [0, 1]], closed=True, facecolor='gray', edgecolor='black', linewidth=2)
ax.add_patch(left_ear)
ax.add_patch(right_ear)

# 3. 繪製眼睛（橢圓形）
left_eye = patches.Ellipse((-0.4, 0.3), 0.3, 0.5, facecolor='white', edgecolor='black', linewidth=2)
right_eye = patches.Ellipse((0.4, 0.3), 0.3, 0.5, facecolor='white', edgecolor='black', linewidth=2)
ax.add_patch(left_eye)
ax.add_patch(right_eye)

# 4. 瞳孔（黑色小圓形）
left_pupil = patches.Circle((-0.4, 0.3), 0.1, facecolor='black')
right_pupil = patches.Circle((0.4, 0.3), 0.1, facecolor='black')
ax.add_patch(left_pupil)
ax.add_patch(right_pupil)

# 5. 鼻子（三角形）
nose = patches.Polygon([[-0.1, 0], [0.1, 0], [0, -0.1]], closed=True, facecolor='pink', edgecolor='black', linewidth=2)
ax.add_patch(nose)

# 6. 嘴巴（弧線）
mouth_left = patches.Arc((-0.05, -0.15), 0.2, 0.2, angle=0, theta1=0, theta2=180, edgecolor='black', linewidth=2)
mouth_right = patches.Arc((0.05, -0.15), 0.2, 0.2, angle=0, theta1=0, theta2=180, edgecolor='black', linewidth=2)
ax.add_patch(mouth_left)
ax.add_patch(mouth_right)

# 7. 鬍鬚（直線）
whiskers = [
    [-0.5, 0, -1.2, 0.3], [-0.5, -0.1, -1.2, -0.1], [-0.5, -0.2, -1.2, -0.5],  # 左邊鬍鬚
    [0.5, 0, 1.2, 0.3], [0.5, -0.1, 1.2, -0.1], [0.5, -0.2, 1.2, -0.5]  # 右邊鬍鬚
]
for x1, y1, x2, y2 in whiskers:
    ax.plot([x1, x2], [y1, y2], color='black', linewidth=2)

# 顯示圖像
plt.show()
```
